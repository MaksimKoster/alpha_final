```python
import pandas as pd

def read_dataset(name):

    train = pd.read_parquet(f'./data/{name}/{name}_train.parquet')
    test = pd.read_parquet(f'./data/{name}/{name}_test.parquet')

    all_data = pd.concat([train, test])

    return all_data
```

```python
names = [
    'fl_credit_card_tendency',
    'invest_prop_4',
    'outflow_12',
    'pd_fl',
    'pd_ul_9',
    'ul_leasing_outflow'
]
```

```python
df = read_dataset(names[0])
```

```python
import os
import requests

import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
import torch

from lightautoml.automl.presets.tabular_presets import TabularAutoML, TabularUtilizedAutoML
from lightautoml.tasks import Task
```

```python
N_THREADS = 16
N_FOLDS = 10
RANDOM_STATE = 42
TEST_SIZE = 0.2
TIMEOUT = 600
TARGET_NAME = 'target'
```

```python
np.random.seed(RANDOM_STATE)
torch.set_num_threads(N_THREADS)
```

```python
train = train.sort_values(by='id')
train
```

```python
print(f'All train {len(train)}')
new_train_size = len(train) - int(len(train) * 0.2)
print(f'New train {new_train_size}')
print(len(train) - new_train_size)
```

```python
new_train = train.head(new_train_size)
new_test = train.tail(len(train) - new_train_size)
```

```python
task = Task('binary')
```

```python
roles = {
    'target': TARGET_NAME,
    'drop': ['smpl','id']
}
```

```python
automl = TabularAutoML(
    task = task, 
    timeout = TIMEOUT,
    cpu_limit = N_THREADS,
    reader_params = {'n_jobs': N_THREADS, 'cv': N_FOLDS, 'random_state': RANDOM_STATE},
)
```

```python
%%time 
out_of_fold_predictions = automl.fit_predict(new_train, roles = roles, verbose = 1)
```

```python
%%time

test_predictions = automl.predict(new_test)
print(f'Prediction for test_data:\n{test_predictions}\nShape = {test_predictions.shape}')
```

```python
#print(f'OOF score: {roc_auc_score(new_train[TARGET_NAME].values, out_of_fold_predictions.data[:, 0])}')
print(f'HOLDOUT score: {roc_auc_score(new_test[TARGET_NAME].values, test_predictions.data[:, 0])}')
```
